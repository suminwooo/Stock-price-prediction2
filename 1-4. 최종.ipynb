### 라이브러리, def

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
def clean(readData):
    text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', readData)
    return text
def intersect(a, b):
    return list(set(a) & set(b))
```

```python
original_data=original_data[['date','open','high','low','close','code']]
original_data=original_data.sort_values(by=['date'])
original_data=original_data.set_index('date')
```

```python
data_01=original_data.loc[start_date_01:finish_date_01]

code_01=[]
for i in data_01['code'].unique():
    code_01.append(i)
    
df_close_01 = pd.DataFrame()
for i in code_01:
    A=data_01[data_01['code']==i]['close']
    df_close_01=pd.concat([df_close_01,A],axis=1)
df_close_01.columns=[i for i in code_01]
COR_01=df_close_01.corr()

cor_01=[]
for i in range(len(COR_01)):
    A_01=COR_01[COR_01.iloc[:,i]>0.70].iloc[:,i].index
    B_01=COR_01[COR_01.iloc[:,i]<0.85].iloc[:,i].index
    AB_01 = (intersect(set(A_01),set(B_01)))
    cor_01.append(['a{}' .format(i) for i in AB_01])
stock_cor_01=[]
for i,j in zip(code_01,cor_01):
    if len(j)>4:
        stock_cor_01.append('a{} = {}' .format(i,j))
    else : 
        continue

split_01=[]
for i in range(len(stock_cor_01)):
    info_01 = stock_cor_01[i].split('=')
    split_01.append(info_01)
    
int_code_01,inter_01=[],[]
for i in split_01:
    int_code_01.append(i[0])
    inter_01.append(clean(i[1]))

df_inter_01=pd.DataFrame()
df_inter_01['code']=int_code_01
df_inter_01['inter']=inter_01
```

### 등분된 각각의 집합을 교집합을 활용

```python
df_inter_01=df_inter_01.set_index('code')
df_inter_02=df_inter_02.set_index('code')
df_inter_03=df_inter_03.set_index('code')
df_inter_04=df_inter_04.set_index('code')
```

```python
df_inter_01.columns=['set1']
df_inter_02.columns=['set2']
df_inter_03.columns=['set3']
df_inter_04.columns=['set4']

final=pd.concat([df_inter_01,df_inter_02,df_inter_03,df_inter_04], axis=1, join='inner')
```

```python
code  = [i for i in final.index]
intersect_code=[]
for i in range(len(final)):
    set1 = final.iloc[i,:][0].strip().replace(' ',',').split(",")
    set2 = final.iloc[i,:][1].strip().replace(' ',',').split(",")
    set3 = final.iloc[i,:][2].strip().replace(' ',',').split(",")
    set4 = final.iloc[i,:][3].strip().replace(' ',',').split(",")
    int_ver1 = intersect(intersect(set1,set2),set3)
    int_ver2 = intersect(intersect(set1,set2),set4)
    int_ver3 = intersect(intersect(set1,set3),set4)
    int_ver4 = intersect(intersect(set2,set3),set4)
    intersect_code.append(list(set(int_ver1+int_ver2+int_ver3+int_ver4)))

final = pd.DataFrame()
final['code'] = code
final['inter'] = intersect_code
final = final.set_index('code')
```

### 데이터 추출하기

```python
code=[]
inter=[]
for i in range(final.shape[0]):
    if len(final['inter'][i]) > 40: # 종목이 90개 이상 겹치는 종목만 뽑기
        code.append(final.index[i])
        inter.append(final.iloc[i,0])
    else:
        continue
```

### 최종 이용할 데이터셋

```python
final_dataset = pd.DataFrame()
final_dataset['code'] = code
final_dataset['interception'] = inter

matrix_code=[]
date_code=[]
code_code=[]

for i in range(len(code)): 
    data = original_data
    data.index = [i for i in original_data.index]
    data = data[data['code']==int(final_dataset['code'][i][1:].strip())]['close'].loc[start_date_01:]

    mean = data.mean(axis=0)
    std = data.std(axis=0)
    data -= mean
    data /= std
    matrix_code.append(list(np.array(data)))
```

```python
date_code.append(data.index)

average_code = pd.DataFrame(sum(np.array(matrix_code))/len(code))
average_code.columns=['close']
average_code.index = [i for i in date_code]
```

```python
plt.figure(figsize=(24, 12))
for i in range(len(code)):
    pd.DataFrame(np.array(matrix_code)[i]).iloc[:,0].plot(label=code[i])

average_code['close'].plot(label='mean',linewidth=7,color='peru')
plt.legend()
plt.show()
```

```python
meet_point = []
original = []
average = []
for i in range(len(code)):
    mean = original_data[original_data['code']==int(code[i][1:].strip())].loc[start_date_01:]['close'].mean()
    std = original_data[original_data['code']==int(code[i][1:].strip())].loc[start_date_01:]['close'].std()
    ori = original_data[original_data['code']==int(code[i][1:])].loc[start_date_01:]['close']
    ave = (average_code['close'] * std ) + mean
    ave.index = [i for i in ori.index]
        
    original.append(ori)
    average.append(ave)
    
    df = pd.DataFrame()
    df['diff'] = ori - ave
    
    count = 0
    for i in range(len(df)-1):
        if df['diff'].values[i] > 0 :
            if df['diff'].values[i+1] < 0 :
                count += 1
            else :
                continue
        if df['diff'].values[i] < 0 :
            if df['diff'].values[i+1] > 0 :
                count += 1
            else :
                continue
    meet_point.append(count)
    
#시작부터 만난 횟수
meet_data = pd.concat([pd.DataFrame(code), pd.DataFrame(meet_point)], axis=1)
meet_data.columns=['code', 'count']
```
